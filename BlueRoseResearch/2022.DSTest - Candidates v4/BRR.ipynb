```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('br_takehome_exam_2022_scoring.csv') # read file
df.head() # view columns/features
df.info() # det missing info
df.describe() # 
# msno.matrix(df) # fill missing values
```

```python
plt.figure(figsize=(10,6))
df[df['good_behavior']==1]['job_aptitude_exam'].hist(alpha=0.5,color='blue',
                                              bins=30,label='Good Behavior = 1')
df[df['good_behavior']==0]['job_aptitude_exam'].hist(alpha=0.5,color='red',
                                              bins=30,label='Good Behavior = 0')
plt.legend()
plt.xlabel('Job Aptitude Exam')
```

```python
plt.figure(figsize=(10,6))
df[df['good_behavior']==1]['unexcused_absences'].hist(alpha=0.5,color='blue',
                                              bins=30,label='Good Behavior = 1')
df[df['good_behavior']==0]['unexcused_absences'].hist(alpha=0.5,color='red',
                                              bins=30,label='Good Behavior = 0')
plt.legend()
plt.xlabel('Unexcused Absences')
```

```python
plt.figure(figsize=(10,6))
df[df['unexcused_absences']==0]['job_aptitude_exam'].hist(alpha=0.5,color='red',
                                              bins=30,label='Absences = 0')
df[df['unexcused_absences']==1]['job_aptitude_exam'].hist(alpha=0.5,color='blue',
                                              bins=30,label='Absences = 1')
df[df['unexcused_absences']==2]['job_aptitude_exam'].hist(alpha=0.5,color='orange',
                                              bins=30,label='Absences = 2')
df[df['unexcused_absences']==3]['job_aptitude_exam'].hist(alpha=0.5,color='green',
                                              bins=30,label='Absences = 3')
plt.legend()
plt.xlabel('Job Aptitude Exam')
plt.ylabel('Absences Count')
```
